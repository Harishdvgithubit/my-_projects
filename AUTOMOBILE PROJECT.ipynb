# # step 1: load data

```python
auto_data=pd.read_csv('d:/auto.txt')
```

```python
auto_data.head()
```

## step 2 : clean the data

```python
'''
    we are going to give this data to machine learning algorithms
    all these alogirthms are written with maths-probability ,stat,calc
    so alogrithm will understand only numbers.
        a.so we should remove symbols.
        b.so we should empty cells.
        c.so we should remove empty cells
        d.so we should remove text/we have to convert text to numbers
        e.we should remove duplicate coloumns and un wanted coloumns
        
'''
```

```python
auto_data = auto_data.replace('?',np.nan)
auto_data.head()
```

```python
auto_data['price'].describe()
```

```python
# we have to convert this column to float type
auto_data['price'] = pd.to_numeric(auto_data['price'], errors='coerce') #coerce means force ignore err
auto_data['price'].describe() #now type is converted to float
```

```python
auto_data = auto_data.drop('normalized-losses', axis=1)
auto_data.head()
```

```python
auto_data['horsepower'].describecribe()
```

```python
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'],errors='coerce')
auto_data['horsepower'].describe()
```

```python
auto_data['num-of-cylinders'].describe()
```

```python
auto_data['num-of-cylinders'].head()
```

```python
cylinders_dict = {
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'eight':8,
    'twelve':12
}
auto_data['num-of-cylinders'].replace(cylinders_dict, inplace = True)
auto_data.head()
```

```python
auto_data['num-of-cylinders'].head()
```

'''
    changing categorical values make, num-of-doors, body-style etc..using one-hot[1-0] representation
'''

```python

auto_data = pd.get_dummies(auto_data, columns=[
    'make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels',
    'engine-location','engine-type','fuel-system'
])
auto_data.head()
```

```python
auto_data.tail()
```

```python
# lets clean up our data
auto_data = auto_data.dropna()
auto_data
```

```python
auto_data.columns
```

```python
# lets see if our data set has any null values?
auto_data[auto_data.isnull().any(axis=1)] #we have only 0 rows..so good..
```

# step 3 :  spilt the data

```python
'''
    total we have 195 data(vehicle details)
    we will spilit this data for training and testing purpose
    
    65% training data & 65% testing data[or]
    70% training data & 30% testing data[or]
    80% training data & 30% testing data[or]
    
    so if we give deatils... we need to find price..
    so price coloumn will be y & remaining coloumns will be x

'''
```

```python
from sklearn.model_selection import train_test_split
# lets feed our data to machine learning model
x = auto_data.drop('price', axis=1)

y = auto_data['price']

# split data into 80% for training, 20% for testing
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=0)
```

## step 4  training the model or alogrithm

'''
 for solving regression problems we will use below alogrithm
 a.linear alogrithm
 b.svr--suporrt vector regression alogrithm
 c.lasso algorithm
 d.ridge alogrithm
  sklearn= sci-kit learn( scientific kit for ml
   sklearn s/w is written with python
'''

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
linear_model.score(x_train, y_train)    # overfitting
```

```python
  predictors = x_train.columns 
coef = pd.Series(linear_model.coef_, predictors).sort_values()
print(coef) 
```

```python
linear_model.coef_
```

# step 5 : testing the model

```python
'''
for testing the alogrithm we have to pass c_test tp predict()function
'''
```

```python
# lets predict using linear regression model
#y_predict is the predicated valued of linear regession alogrithm
y_predict = linear_model.predict(x_test)
```
